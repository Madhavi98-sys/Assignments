> Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
#CODE HERE
multiples7 = []
multiples5 = []
for i in range(2000,3200):
    if i % 7 == 0:
        multiples7.append(i)
        
        if i % 5 ==0:
            multiples5.append(i)
for i in multiples7:
    if i not in multiples5:
        print(i,end = ",")
```

> Write a program which can compute the factorial of a given numbers.
The results should be printed in a comma-separated sequence on a single line.
Suppose the following input is supplied to the program:
8
Then, the output should be:
40320

```python

def fact(n):
  if n == 0:
    return 1
  else:
    return n * fact(n - 1)

n = int(input("Enter the value to compute the factorial:"))
print(fact(n))
print(f"{n}!={fact(n)}")
```

> With a given integral number n, write a program to generate a dictionary that contains (i, i*i) such that is an integral number between 1 and n (both included). and then the program should print the dictionary.

8

{1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64}

> create user defined fuction for 
- addition "add"
- subtraction "sub"
- multiplication "mul"
- division "div"
- power "pow"

```python
def add(n1,n2):
  return n1+n2
def sub(n1,n2):
    return n1-n2
def mul(n1,n2):
    return n1 *n2
def div(n1,n2):
    return n1 % n2
def pow(n1,n2):
    return n1** n2

n1 = int(input("Please enter the value : "))
n2 = int(input("Please enter the value : "))
k = input("Enter the keyword {add, sub, mul, div, pow}:")
if k in globals():
    result = globals()[k](n1, n2)
    print(result)
else:
    print("Invalid keyword")
```

### [1,1,1,1,3,3,4,4,4,5,6,7,8,8,8]
#### 1. from the above list print the uniques values
#### 2. count how many times each number exist
#### 3. Print the number which is occured more number of times

```python
list1 = [1, 1, 1, 1, 3, 3, 4, 4, 4, 5, 6, 7, 8, 8, 8]
myset = set(list1)
print("Unique values are:", myset)
dic = {}
# Count how many times each number exists
for i in myset:
    count = 0
    j = 0
    while j < len(list1):
        if i == list1[j]:
            count += 1
        j += 1
    dic.update({i:count})
print(dic)
#Print the number which is occured more number of times
max_key = max(dic, key=dic.get)
print(f"The number that occurs the most times is {max_key}, with {dic[max_key]} occurrences.")
```
